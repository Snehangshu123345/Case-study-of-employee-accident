```python
# function to create season 
def season_creation(x):
    if x in [9,10,11]:
        season="Spring"
    elif x in [12,1,2]:
        season="Summer"
    elif x in [3,4,5]:
        season="Autum"
    elif x in [6,7,8]:
        season="Winter"
    return season
```

#NLP preprocessing
Description column contains the details why accidents happend.So we tried to add new features by using this import information
with NLP technique.

Stopwords-->


```python
from sklearn.feature_extraction.text import TfidfVectorizer
#Pip install nltk
from nltk import tokenize,stem
import nltk
nltk.download("vader_lexicon")
# Sentiment intensity analyzer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
#pip install wordcloud
from wordcloud import WordCloud,STOPWORDS
```

```python
STOPWORDS.update(["kg","cm","am","pm","ore","mr","mrs"])
print(STOPWORDS)
```

```python
# NLP preprocessing 
```

```python
def nlp_preprocessor(row):
    sentence=row.Description
    #convert all characters to lowercase
    lowered=sentence.lower()
    tok=tokenize.word_tokenize(lowered)
    
    #Lematization & stemming
    lemmatizer=stem.WordNetLemmatizer()
    lem=[lemmatizer.lemmatize(i) for i in tok if i not in STOPWORDS]
    stemmer=stem.PorterStemmer()
    stems=[stemmer.stem(i) for i in lem if i not in STOPWORDS]
    
    # remove non-alphabetical characters like(":",".",",")
    alphas=[i for i in stems if i.isalpha() and (i not in STOPWORDS)]
    return " ".join(alphas)
```

```python
# we are converting the description column -- lower casing,stemming,Lemmatization...
```

```python
df["Description_processed"]=df.apply(nlp_preprocessor,axis=1)
df.head(2)
```

```python
# calculate sentiment intensity analyzer
```

```python
def sentiment_calculate(text):
    analyzer=SentimentIntensityAnalyzer()
    sent_score=analyzer.polarity_scores(text)["compound"]
    
    return float(sent_score)
```

```python
df["Decription_sentiment_score"]=df["Description_processed"].apply(lambda x: sentiment_calculate(x) )
```
